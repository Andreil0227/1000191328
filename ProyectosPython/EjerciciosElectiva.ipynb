```python
# Ejercicios para el buen manejo de pandas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Punto 1 con numpy

```python
# Crear lista usando np.arange para generar la lista del 1 al 10
numeros1_10= np.arange(1, 11)
print (numeros1_10)
```

```python
# Suma + 5 de la lista
suma_numeros = numeros1_10 + 5
print (f"La suma de los números es_ {suma_numeros}")
```

```python
# Multiplicación del array
multiplicacion_numeros = numeros1_10 * 5
multiplicacion_suma = suma_numeros * 5
print (f"La multiplicación fue: {multiplicacion_numeros}")
print (f"La multiplicación de la suma fue: {multiplicacion_suma}")
```

```python
# Transpuesta de todas las matrices creadas.
transpuesta = np.array([numeros1_10 ,suma_numeros, multiplicacion_numeros, multiplicacion_suma])
print ("La transpuesta de todas las matrices es: \n ", transpuesta.T)
```

# Dataframe

```python
# Dataframe con Pandas
data = {'Nombre' : ['Juan','María','Fernanda','Jose', 'Giver'],
        'Edad':[30,45,29,30, 15],
        'Salario': [2500, 4000, 3500, 6000, 8500]}

df = pd.DataFrame(data)
print("Dataframe: \n", df)
```

```python
# Agragar columnda de bonificación

df['Bonificacion'] = df['Salario']*0.15
print ("Tabla con bonificación \n", df)
```

```python
# Personas con un salario mayor de 3500

print (df[df['Salario'] > 3500])
```

```python
# Estadisticas generales

print (df.describe())
```

# Matplotlib  

```python
# Linea recta en matplot poniendo etiquetas.

x = [0, 10]
y = [5, 5]

plt.plot(x, y, label='linea recta', color = 'red', )
plt.title (" Ejercicio Matplotlib ")
plt.xlabel ("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid()
```

# Seaborn

```python
# Sacamos el dataset

Pinguinos = sns.load_dataset("penguins")
```

```python
# Mostramos las primeras 5 filas

PrimerosCinco = Pinguinos.head()
print(PrimerosCinco)
```

```python
# Creación del grafico
plt.figure(figsize=(8, 6))
sns.violinplot(x="species", y="body_mass_g", hue="sex", data=Pinguinos, split=True)


title = "Distribución del Peso Corporal de los Pingüinos por Especie y Sexo"
plt.title(title)


plt.show()
```
